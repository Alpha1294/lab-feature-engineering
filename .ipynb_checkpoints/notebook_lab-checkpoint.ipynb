## Categorical Variables

Here we will try to reduce the number of categories. An ideal way would have been to group the states into 
regions. But in this case we will group all the states with counts less than 2500 into one category "other"

```python
pd.DataFrame(categorical['STATE'].value_counts()).reset_index()
```

```python
df = pd.DataFrame(categorical['STATE'].value_counts()).reset_index()

df.columns = ['state', 'count']
df
other_states = list(df[df['count']<2500]['state'])
other_states

def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
categorical['STATE'] = list(map(clean_state, categorical['STATE']))
```

# Lab | Handling Data Imbalance in Classification Models

**Begin the Modeling here**
- Look critically at the dtypes of numerical and categorical columns and make changes where appropriate.
- Concatenate numerical and categorical back together again for your X dataframe.  Designate the TargetB as y.
  - Split the data into a training set and a test set.
  - Split further into train_num and train_cat.  Also test_num and test_cat.
  - Scale the features either by using MinMax Scaler or a Standard Scaler. (train_num, test_num)
  - Encode the categorical features using One-Hot Encoding or Ordinal Encoding.  (train_cat, test_cat)
      - **fit** only on train data, transform both train and test
      - again re-concatenate train_num and train_cat as X_train as well as test_num and test_cat as X_test
  - Fit a logistic regression model on the training data.
  - Check the accuracy on the test data.

**Note**: So far we have not balanced the data.

Managing imbalance in the dataset

- Check for the imbalance.
- Use the resampling strategies used in class for upsampling and downsampling to create a balance between the two classes.
- Each time fit the model and see how the accuracy of the model has changed.

```python
# Concatenate numerical and categorical back together again for your X dataframe. Designate the TargetB as ycategorical
logistic_data = pd.concat([numerical,categorical], axis = 1)
logistic_data
```

```python
# that lets my logstic_data df as the features
#Split further into train_num and train_cat. Also test_num and test_cat.
X = logistic_data
X_train, X_test, y_train, y_test = train_test_split(X, y_target_b, random_state=0) # default fraction is .2

#splitting into categorical and numerical
X_train_num = X_train.select_dtypes(include = np.number)
X_train_cat = X_train.select_dtypes(include = object)

X_test_num = X_test.select_dtypes(include = np.number)
X_test_cat = X_test.select_dtypes(include = object)
```

```python
numerical["WEALTH1"].value_counts()
```

```python
X_train_cat
```

```python
# Scale the features either by using MinMax Scaler or a Standard Scaler. (train_num, test_num)
transformer = MinMaxScaler().fit(X_train_num) # fit with the train data
numericals_train_scaled = transformer.transform(X_train_num)
numericals_test_scaled = transformer.transform(X_test_num)
```

```python
#X_train_cat.applymap(str)
# X_train_cat = X_train_cat.drop(["WEALTH1","WEALTH2"],axis=1)
# X_test_cat =  X_test_cat.drop(["WEALTH1","WEALTH2"],axis=1)
```

```python
#encode categoricals
encoder = OneHotEncoder() # fit only train data
#fit and transform
encoded_data = encoder.fit(X_train_cat)
categoricals_train_encoded = encoder.transform(X_train_cat).toarray()
categoricals_test_encoded = encoder.transform(X_test_cat).toarray()
```

```python
X_train_transformed = np.concatenate((numericals_train_scaled,categoricals_train_encoded),axis=1)
X_test_transformed = np.concatenate((numericals_test_scaled,categoricals_test_encoded),axis=1)
```

### building a logistic regression model

```python
LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_transformed, y_train)
LR.score(X_test_transformed, y_test)   
```

```python
pred = LR.predict(X_test_transformed)
confusion_matrix(y_test,pred)
```

```python
precision = precision_score(y_test, pred, zero_division=0)
recall = recall_score(y_test, pred, zero_division=0)
f1 = f1_score(y_test, pred, zero_division=0)

print("precision: ", precision)
print("recall: ", recall)
print("F1: ", f1)
```

```python
# now thats interesting, since im not predicting any positive cases (true positivies) im getting 0 score, i gues is because of the imbalance, so lets balance 
X_train, X_test, y_train, y_test = train_test_split(X, y_target_b, random_state=0) # default fraction is .2
X_train_imb = pd.DataFrame(X_train)
trainset = pd.concat((X_train, y_train), axis=1)
```

```python
trainset['TARGET_B'].value_counts()
```

```python
category_0 = trainset[trainset['TARGET_B'] == 0]
category_1 = trainset[trainset['TARGET_B'] == 1]
```

```python
category_0_undersampled = resample(category_0, 
                                   replace=False, 
                                   n_samples = len(category_1))
```

```python
print(category_0_undersampled.shape)
print(category_1.shape)
```

```python
train_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
train_downsampled['TARGET_B'].value_counts()
```

```python
X_train = train_downsampled.drop('TARGET_B',axis = 1)
y_train = train_downsampled['TARGET_B']
```

```python
#splitting into categorical and numerical
X_train_num = X_train.select_dtypes(include = np.number)
X_train_cat = X_train.select_dtypes(include = object)

X_test_num = X_test.select_dtypes(include = np.number)
X_test_cat = X_test.select_dtypes(include = object)

transformer = MinMaxScaler().fit(X_train_num) 
numericals_train_scaled = transformer.transform(X_train_num)
numericals_test_scaled = transformer.transform(X_test_num)

encoder = OneHotEncoder(handle_unknown='error', drop='first').fit(X_train_cat)
categoricals_train_encoded = encoder.transform(X_train_cat).toarray()
categoricals_test_encoded = encoder.transform(X_test_cat).toarray()

X_train_transformed = np.concatenate((numericals_train_scaled,categoricals_train_encoded),axis=1)
X_test_transformed = np.concatenate((numericals_test_scaled,categoricals_test_encoded),axis=1)

LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_transformed, y_train)
LR.score(X_test_transformed, y_test)   
```

```python
pred = LR.predict(X_test_transformed)
confusion_matrix(y_test,pred)
```

```python
#much better now!
precision = precision_score(y_test, pred, zero_division=0)
recall = recall_score(y_test, pred, zero_division=0)
f1 = f1_score(y_test, pred, zero_division=0)

print("precision: ", precision)
print("recall: ", recall)
print("F1: ", f1)
```
